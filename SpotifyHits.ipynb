```python
import numpy as np
import matplotlib.pyplot as plt
from keras import Sequential
from keras.layers import Dense
from keras.initializers import glorot_normal
import pandas as pd
import random
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
music = pd.read_csv("dataSpotify.csv")
music = music.drop(['time_signature','keis','duration_ms'],axis=1)
music.head()
```

```python
rows = music.values.shape[1]
print(rows)

# rows80 = int(rows*0.9)
# rows20 = int(rows*0.1)
X = np.array(music.iloc[: ,0:10].values, dtype=float)
Y = np.array(music.iloc[: ,12:].values, dtype=float)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=73)
validate = pd.read_csv("dataSpotify.csv")
validate = validate.drop(['speechiness','acousticness','danceability','duration_ms','energy', 'instrumentalness', 'keis', 'liveness', 'loudness', 'mode', 'tempo', 'time_signature', 'valence'],axis=1)
```

```python
out_dimension = 1
nHidden = 15
nInputDimensions = X_train.shape[1]
print(nInputDimensions)
```

```python
modelo = Sequential()
modelo.add(Dense(12, input_dim = nInputDimensions, activation='sigmoid', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(15, activation='elu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(17, activation='elu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(15, activation='elu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(12, activation='elu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(5, activation='elu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(2, activation='elu', kernel_initializer = glorot_normal(seed=0)))
modelo.add(Dense(out_dimension, activation='sigmoid', kernel_initializer = glorot_normal(seed=0)))
modelo.compile(loss='mean_squared_error',optimizer='adam', metrics=['accuracy'])
training= modelo.fit(X_train,Y_train,epochs=300)
```

```python
resultados= modelo.evaluate(X_test, Y_test, batch_size=len(Y_test))
print('Accuracy',resultados[1])
```

```python
predicciones = modelo.predict(X)
predicciones
```

```python
pre = pd.Series(predicciones)
validate['prediccion'] = pre
validate['prediccion'] = validate['prediccion'].str.get(0)
validate
```

```python
coincidencias = []
for dato in validate.prediccion:
    if dato >= 0.5:
        coincidencias.append(1)
    else:
        coincidencias.append(0)
validate['final'] = coincidencias
validate
```
